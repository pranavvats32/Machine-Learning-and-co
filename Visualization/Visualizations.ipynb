# Plots

```python
import matplotlib.pyplot as plt
import numpy as np
```

## Scatter Plots

```python
X = np.random.rand(1000, 2)
plt.scatter(X[:, 0], X[:, 1])
plt.show()
```

## Bar Plots, Box Plots & Histograms

```python
X = np.random.rand(5)
plt.bar(np.arange(5), X)
plt.show()
```

```python
data_1 = np.random.rand(100)
data_2 = np.random.rand(100) * 2 + 1
data_3 = np.random.rand(100) * 2 - 1
plt.boxplot([data_1, data_2, data_3])
plt.show()
```

```python
data = np.random.rand(1000)
plt.hist(data, bins=5)
plt.show()
```

```python
plt.hist(data, bins=np.linspace(0, 1, 100))
plt.show()
```

```python
plt.hist(data, bins="fd")
plt.show()
```

## Line Plots

```python
x = np.linspace(-np.pi, np.pi, 1000, endpoint=True)
c = np.cos(x)
s = np.sin(x)
```

```python
plt.plot(x, c)
plt.show()
```

## Modifying Plots

```python
plt.plot(x, c)
plt.plot(x, s)
plt.show()
```

```python
plt.plot(x, c)
plt.plot(x, s, color="red")
plt.show()
```

```python
plt.plot(x, c)
plt.plot(x, s, color="red")
plt.ylim(-1, 1)
plt.ylabel("f(x)")
plt.xlabel("x")
plt.show()
```

```python
plt.plot(x, c)
plt.plot(x, s, color="red")
plt.xticks([-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi])
plt.yticks([-1, 0, 1])
plt.show()
```

```python
plt.plot(x, c)
plt.plot(x, s, color="red")
plt.xticks(
    [-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
    [r"$-\pi$", r"$-\pi/2$", r"$0$", r"$+\pi/2$", r"$+\pi$"],
)
plt.yticks([-1, 0, 1])
plt.show()
```

```python
plt.plot(x, c, label="cosine")
plt.plot(x, s, color="red", label="sine")
plt.xticks(
    [-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
    [r"$-\pi$", r"$-\pi/2$", r"$0$", r"$+\pi/2$", r"$+\pi$"],
)
plt.yticks([-1, 0, 1])
plt.legend(loc="upper left", frameon=False)
plt.show()
```

```python
plt.plot(x, c, label="cosine")
plt.plot(x, s, color="red", label="sine")
plt.xticks(
    [-np.pi, -np.pi / 2, 0, np.pi / 2, np.pi],
    [r"$-\pi$", r"$-\pi/2$", r"$0$", r"$+\pi/2$", r"$+\pi$"],
)
plt.yticks([-1, 0, 1])
plt.legend(loc="upper left", frameon=False)

ax = plt.gca()  # get current axis
ax.spines["right"].set_color("none")
ax.spines["top"].set_color("none")
ax.spines["left"].set_position("center")
ax.spines["bottom"].set_position("center")
plt.show()
```

# Parallel Coordinates

```python
import pandas as pd
from sklearn.datasets import load_iris
import matplotlib.pyplot as plt
```

```python
# load iris dataset
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)
df["target"] = iris.target
target_names = {0: "setosa", 1: "versicolor", 2: "virginica"}
df["target_names"] = df["target"].map(target_names)

# plot parallel lines
pd.plotting.parallel_coordinates(df, "target_names")
plt.show()
```
